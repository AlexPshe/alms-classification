```python
# parser to cvs
import lxml.html
import lxml.etree
import requests
import pandas as pd
import numpy as np
```

```python
website = 'http://www.vybory.izbirkom.ru/region/region/izbirkom?action=show&root=1&tvd=100100084849066&vrn=100100084849062&region=0&global=1&sub_region=0&prver=0&pronetvd=null&vibid=100100084849066&type=227'
response = requests.get(website)
tree = lxml.html.fromstring(response.text)
links = tree.xpath('//option/@value')
```

```python
region_link_example = tree.xpath('//option/@value')[0]
tree_region_ex = lxml.html.fromstring(requests.get(region_link_example).text)
name_column = tree_region_ex.xpath('//table/tr/td[contains(@align,"left")]/nobr/text()')
name_column.insert(0, 'Номер УИК')
name_column.insert(0, 'Название ТИК')
name_column.insert(0, 'Регион')

df = pd.DataFrame()
```

```python
count = 0
for link in links:
    tree = lxml.html.fromstring(requests.get(link).text)
    region_links = tree.xpath('//option/@value')
    for region_link in region_links:
        
        tree_region = lxml.html.fromstring(requests.get(region_link).text)
        
        page_link = tree_region.xpath('//table/tr/td/a/@href')[3]
        tree_comm = lxml.html.fromstring(requests.get(page_link).text)
        
        region = (tree_region.xpath('//table/tr/td/a/text()')[1])
        name_tik = (tree_region.xpath('//table/tr/td/a/text()')[2])
        
        count = count + 1
        print(count, "=  region:", region, " name_tik: ", name_tik)
        
        all_data = tree_comm.xpath('//table/tr/td[contains(@align,"right")]/nobr/b/text()')[20:]
        numbers_yik = tree_comm.xpath('//table/tr[contains(@valign, "top")]/td[contains(@align,"center")]/nobr/text()')
        
        df_local = pd.DataFrame(np.array(all_data).reshape(20, len(all_data)//20).transpose(), columns=range(3,23))
        
        region_column = np.full((len(numbers_yik), 1), region)
        name_tik_column = np.full((len(numbers_yik), 1), name_tik)
        
        df_local.insert(loc=0, value=numbers_yik, column = '2')
        df_local.insert(loc=0, value=name_tik_column, column = '1')
        df_local.insert(loc=0, value=region_column, column = '0')
        
        df = df.append(df_local, ignore_index=True)
```

```python
df.columns = name_column
df
```

```python
df.to_csv('data_results_vibori.csv', sep='\t', encoding='utf-8')
```
